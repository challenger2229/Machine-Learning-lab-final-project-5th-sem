```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv('/content/Copy of sonar data.csv', header=None)
```

```python
df.columns = [f'Feature_{i}' for i in range(1, 61)] + ['Target']
df.head()
```

this dataset has been retrived from kaggle and it is about sonar readings of various objects including underwater mines also , this dataset has 207 rows and 60 columns. therefore for understanding it in project we need to visualise it here itself

since 60 features are too big to be plotted , we are using first 10 features for plotting for **pair plotting**

```python


sns.pairplot(df.iloc[:, :10])
plt.show()
```

**Correlation Heatmap:** This will show correlations between features, which is useful for understanding feature relationships.

```python
plt.figure(figsize=(15, 12))
sns.heatmap(df.iloc[:, :-1].corr(), cmap='coolwarm', annot=False)
plt.title('Feature Correlation Heatmap')
plt.show()
```

**Box Plot:** To view distributions, especially if some features are outliers.

```python
plt.figure(figsize=(18, 8))
sns.boxplot(data=df.iloc[:, :-1])
plt.xticks(rotation=90)
plt.title('Boxplot of All Features')
plt.show()
```
